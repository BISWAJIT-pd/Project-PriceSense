```python
df['timestamp']=pd.to_datetime(df['timestamp'])
df['date']=df['timestamp'].dt.date
df['time']=df['timestamp'].dt.strftime('%H:%M')
```

```python
df.sample(5)
```

```python
df.drop(columns='timestamp',inplace=True)
```

```python
df.sample(5)
```

```python
df.describe()
```

```python
def neg(x):
  if(x<0):
    return x*(-1)
  else:
    return x
df['price']=df['price'].apply(neg)
df['quantity']=df['quantity'].apply(neg)
```

```python
df.describe()
```

```python
df.info()
```

```python
trans=df
```

```python
merge_1= pd.merge(trans, product, on='product_id', how='left')
```

```python
merge_1.sample(5)
```

```python
merge_2= pd.merge(merge_1, geo, on='order_id', how='left')
```

```python
merge_2.info()
```

```python
merge_2[merge_2['state'].isnull()].sample(5)
```

```python
merge_2['state'].value_counts()
```

```python
merge_2['occasion'].value_counts()
```

```python
#I WANT TO IMPUTE MISSING VALUES AT RANDOM IN COLUMNS SO THAT DATA DISTRIBUTION REMAIN SAME
states = ['New York', 'California', 'Illinois', 'Colorado',
          'Washington', 'Texas', 'Florida', 'Georgia']
merge_2.loc[merge_2['state'].isna(), 'state'] = np.random.choice(
    states,
    size=merge_2['state'].isna().sum()
)
```

```python
occasions = ['late-night', 'religious-fasting', 'on-the-go', 'festive',
          'daily snack', 'gym', 'road-trip', 'marathon-prep']
merge_2.loc[merge_2['occasion'].isna(), 'occasion'] = np.random.choice(
    occasions,
    size=merge_2['occasion'].isna().sum()
)
tie = ['Tier 1', 'Tier 2', 'Tier 3', 'Unknown']
merge_2.loc[merge_2['city_tier'].isna(), 'city_tier'] = np.random.choice(
    tie,
    size=merge_2['city_tier'].isna().sum()
)
```

```python
merge_2.info()
```

```python
#MERGE ALL OF THEM TO A SINGLE DATASET
df=pd.merge(merge_2, con, on='user_id', how='left')
```
